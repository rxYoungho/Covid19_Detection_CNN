```
import numpy as np
import time
import pandas as pd

import PIL.Image as Image
import matplotlib.pylab as plt

import tensorflow as tf
import tensorflow_hub as hub

from sklearn.metrics import confusion_matrix



def plot_confusion_matrix(cm,
                          target_names,
                          title='Confusion matrix',
                          cmap=None,
                          normalize=True):
    """
    given a sklearn confusion matrix (cm), make a nice plot

    Arguments
    ---------
    cm:           confusion matrix from sklearn.metrics.confusion_matrix

    target_names: given classification classes such as [0, 1, 2]
                  the class names, for example: ['high', 'medium', 'low']

    title:        the text to display at the top of the matrix

    cmap:         the gradient of the values displayed from matplotlib.pyplot.cm
                  see http://matplotlib.org/examples/color/colormaps_reference.html
                  plt.get_cmap('jet') or plt.cm.Blues

    normalize:    If False, plot the raw numbers
                  If True, plot the proportions

    Usage
    -----
    plot_confusion_matrix(cm           = cm,                  # confusion matrix created by
                                                              # sklearn.metrics.confusion_matrix
                          normalize    = True,                # show proportions
                          target_names = y_labels_vals,       # list of names of the classes
                          title        = best_estimator_name) # title of graph

    Citiation
    ---------
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html

    """
    import itertools

    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy

    if cmap is None:
        cmap = plt.get_cmap('Blues')

    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names, rotation=45)
        plt.yticks(tick_marks, target_names)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]


    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")


    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label\naccuracy={:0.4f}; misclass={:0.4f}'.format(accuracy, misclass))
    plt.show()

def fix_dims(tens):
  return np.repeat(tens, 3, -1)

IMAGE_SHAPE = (224, 224)

NUM_CLASSES = 3

EPOCH_NUM = 20

pd.set_option('display.max_colwidth',1000)

"""
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)#, preprocessing_function=fix_dims)
image_data = image_generator.flow_from_directory("CT-Data", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=True)#grayscale
image_generator2 = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)#, preprocessing_function=fix_dims)
val_data = image_generator2.flow_from_directory("CT-Data-Youngho", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=False)
"""


image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255, validation_split=0.25)#, preprocessing_function=fix_dims)
image_data = image_generator.flow_from_directory("Xray-Data", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=True, class_mode= 'sparse', subset='training')#grayscale

val_data = image_generator.flow_from_directory("Xray-Data", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=True, class_mode= 'sparse', subset='validation')


imgen1 = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
imgdata1 = imgen1.flow_from_directory("Xray-Data", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=True, class_mode= 'sparse', classes=["COVID19", "NORMAL"])

imgen2 = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
imgdata2 = imgen1.flow_from_directory("Xray-Data", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=True, class_mode= 'sparse', classes=["NORMAL", "PNEUMONIA"])

imgen3 = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)
imgdata3 = imgen1.flow_from_directory("Xray-Data", target_size=IMAGE_SHAPE, color_mode="rgb", batch_size=128, shuffle=True, class_mode= 'sparse', classes=["COVID19", "PNEUMONIA"])


for image_batch, label_batch in image_data:
  print("Image batch shape: ", image_batch.shape)
  print("Label batch shape: ", label_batch.shape)
  print("Label batch data: ", label_batch)
  break

#{'COVID19': 0, 'NORMAL': 1, 'PNEUMONIA': 2}
mapdict = image_data.class_indices
print("class mapping:", mapdict)

def pneum(datagen):
  label_change = np.array([0, 0, 1])
  for data, label in datagen:
    label = label_change[label.astype(int)].astype(float)
    yield data, label

def covid(datagen):
  label_change = np.array([1, 0, 0])
  for data, label in datagen:
    for idx in range(len(label)):
      label = label_change[label.astype(int)].astype(float)
    yield data, label

def normal(datagen):
  label_change = np.array([0, 1, 0])
  for data, label in datagen:
    for idx in range(len(label)):
      label = label_change[label.astype(int)].astype(float)
    yield data, label

print(image_data.class_indices)

first_train = True

ckpt_dir1 = "ckpt1/cp.ckpt"

ckpt_dir2 = "ckpt2/cp.ckpt"

ckpt_dir3 = "ckpt3/cp.ckpt"

#feature_extractor_layer = hub.KerasLayer("https://tfhub.dev/google/imagenet/pnasnet_large/classification/4")

img_inputs = tf.keras.Input(shape=(224, 224, 3))

img_inputs.trainable = False

"""
print("Downloading pre-trained Resnet50 #1")

feature_extractor_layer1 = hub.KerasLayer("https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/4",
              trainable=False, input_shape=(224, 224, 3))(img_inputs) #, arguments=dict(batch_norm_momentum=0.997)

print("Downloading pre-trained Resnet50 #2")

feature_extractor_layer2 = hub.KerasLayer("https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/4",
              trainable=False, input_shape=(224, 224, 3))(img_inputs) #, arguments=dict(batch_norm_momentum=0.997)

print("Downloading pre-trained Resnet50 #3")

feature_extractor_layer3 = hub.KerasLayer("https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/4",
              trainable=False, input_shape=(224, 224, 3))(img_inputs) #, arguments=dict(batch_norm_momentum=0.997)
"""

print("Loading pre-trained Resnet50 #1")

feature_extractor_layer1 = tf.keras.applications.ResNet50(include_top=False, weights='imagenet', input_shape=(224, 224, 3), pooling = 'avg')
feature_extractor_layer1.trainable = False
feature_extractor_layer1._name = "resnet1"
feature_extractor_layer1 = feature_extractor_layer1(img_inputs)

print("Loading pre-trained Resnet50 #2")

feature_extractor_layer2 = tf.keras.applications.ResNet50(include_top=False, weights='imagenet', input_shape=(224, 224, 3), pooling = 'avg')
feature_extractor_layer2.trainable = False
feature_extractor_layer2._name = "resnet2"
feature_extractor_layer2 = feature_extractor_layer2(img_inputs)

print("Loading pre-trained Resnet50 #3")

feature_extractor_layer3 = tf.keras.applications.ResNet50(include_top=False, weights='imagenet', input_shape=(224, 224, 3), pooling = 'avg')
feature_extractor_layer3.trainable = False
feature_extractor_layer3._name = "resnet3"
feature_extractor_layer3 = feature_extractor_layer3(img_inputs)

#Add dropout
dropout1 = tf.keras.layers.Dropout(0.5)(feature_extractor_layer1)

dropout2 = tf.keras.layers.Dropout(0.5)(feature_extractor_layer2)

dropout3 = tf.keras.layers.Dropout(0.5)(feature_extractor_layer3)

#Make 3 ResNet50 models

dense1 = tf.keras.layers.Dense(1, activation="sigmoid")(dropout1)

dense2 = tf.keras.layers.Dense(1, activation="sigmoid")(dropout2)

dense3 = tf.keras.layers.Dense(1, activation="sigmoid")(dropout3)

#Wrap each in a dense layer for training

#Package each Resnet separately for binary classification
resnet1 = tf.keras.Model(inputs=img_inputs, outputs=dense1, name="resnet1")

resnet2 = tf.keras.Model(inputs=img_inputs, outputs=dense2, name="resnet2")

resnet3 = tf.keras.Model(inputs=img_inputs, outputs=dense3, name="resnet3")

#COMPLETELY NEW BITS
#CONCATENATE DENSE LAYERS; THEN ADD RELU CHAIN (which will be trained as the last training step)

concat = tf.keras.layers.Concatenate()([dense1, dense2, dense3])

concat.trainable = False

flatten = tf.keras.layers.Flatten()(concat)

flatten.trainable = False

relu1 = tf.keras.layers.Dense(units=512, activation='tanh')(concat)

relu2 = tf.keras.layers.Dense(units=256, activation='tanh')(relu1)

relu3 = tf.keras.layers.Dense(units=192, activation='tanh')(relu2)

relu4 = tf.keras.layers.Dense(units=128, activation='tanh')(relu3)

output = tf.keras.layers.Dense(3, activation="softmax")(concat)

model = tf.keras.Model(inputs=img_inputs, outputs=output, name="triple-resnet")

model.summary()



tf.keras.utils.plot_model(model, "triple-resnet-structure.png", show_shapes=True)

class CollectBatchStats(tf.keras.callbacks.Callback):
  def __init__(self):
    self.batch_losses = []
    self.batch_acc = []

  def on_train_batch_end(self, batch, logs=None):
    self.batch_losses.append(logs['loss'])
    self.batch_acc.append(logs['acc'])
    self.model.reset_metrics()

print("Training 1: dense1 / Pneumonia (everything else is frozen)")

for image_batch, label_batch in imgdata1:
  print("Image batch shape: ", image_batch.shape)
  print("Label batch shape: ", label_batch.shape)
  print("Label batch data: ", label_batch)
  break

dense1.trainable = True

dense2.trainable = False

dense3.trainable = False

relu1.trainable = False

relu2.trainable = False

relu3.trainable = False

relu4.trainable = False

output.trainable = False

resnet1.summary()

resnet1.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
  metrics=['acc'])


steps_per_epoch = np.ceil(imgdata1.samples/imgdata1.batch_size)

batch_stats_callback = CollectBatchStats()

history = resnet1.fit(imgdata1, epochs=EPOCH_NUM,
                    steps_per_epoch=steps_per_epoch,
                    callbacks=[batch_stats_callback])

plt.plot(history.history['acc'])
plt.title('resnet1 accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.title('resnet1 loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

print("Training 2: dense2 / COVID (everything else is frozen)")

for image_batch, label_batch in imgdata2:
  print("Image batch shape: ", image_batch.shape)
  print("Label batch shape: ", label_batch.shape)
  print("Label batch data: ", label_batch)
  break

dense1.trainable = False

dense2.trainable = True

dense3.trainable = False

relu1.trainable = False

relu2.trainable = False

relu3.trainable = False

relu4.trainable = False

output.trainable = False

resnet2.summary()

resnet2.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
  metrics=['acc'])


steps_per_epoch = np.ceil(imgdata2.samples/imgdata2.batch_size)

batch_stats_callback = CollectBatchStats()

history = resnet2.fit(imgdata2, epochs=EPOCH_NUM,
                    steps_per_epoch=steps_per_epoch,
                    callbacks=[batch_stats_callback])

plt.plot(history.history['acc'])
plt.title('resnet2 accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.title('resnet2 loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

print("Training 3: dense3 / Regular (everything else is frozen)")

for image_batch, label_batch in imgdata3:
  print("Image batch shape: ", image_batch.shape)
  print("Label batch shape: ", label_batch.shape)
  print("Label batch data: ", label_batch)
  break

dense1.trainable = False

dense2.trainable = False

dense3.trainable = True

relu1.trainable = False

relu2.trainable = False

relu3.trainable = False

relu4.trainable = False

output.trainable = False

resnet3.summary()

resnet3.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
  metrics=['acc'])


steps_per_epoch = np.ceil(imgdata3.samples/imgdata3.batch_size)

batch_stats_callback = CollectBatchStats()

history = resnet3.fit(imgdata3, epochs=EPOCH_NUM,
                    steps_per_epoch=steps_per_epoch,
                    callbacks=[batch_stats_callback])

plt.plot(history.history['acc'])
plt.title('resnet3 accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.title('resnet3 loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

print("Training 4: relu1~relu4 + output / Full dataset (everything else is frozen)")

for image_batch, label_batch in image_data:
  print("Image batch shape: ", image_batch.shape)
  print("Label batch shape: ", label_batch.shape)
  print("Label batch data: ", label_batch)
  break

dense1.trainable = False

dense2.trainable = False

dense3.trainable = False

relu1.trainable = True

relu2.trainable = True

relu3.trainable = True

relu4.trainable = True

output.trainable = True

"""resnet1.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
  metrics=['acc'])

resnet2.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
  metrics=['acc'])

resnet3.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
  metrics=['acc'])"""

model.summary()

model.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['acc'])


steps_per_epoch = np.ceil(image_data.samples/image_data.batch_size)

batch_stats_callback = CollectBatchStats()

history = model.fit(image_data, epochs=EPOCH_NUM,
                    steps_per_epoch=steps_per_epoch,
                    callbacks=[batch_stats_callback])

plt.plot(history.history['acc'])
plt.title('full model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.title('full model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()




label_id = np.argmax(label_batch, axis=-1)

class_names = sorted(image_data.class_indices.items(), key=lambda pair:pair[1])
class_names = np.array([key.title() for key, value in class_names])
predicted_batch = model.predict(val_data)
predicted_id = np.argmax(predicted_batch, axis=-1)
predicted_label_batch = class_names[predicted_id]





"""
plt.figure(figsize=(10,9))
plt.subplots_adjust(hspace=0.5)
for n in range(30):
  plt.subplot(6,5,n+1)
  plt.imshow(image_batch[n])
  color = "green" if predicted_id[n] == label_id[n] else "red"
  plt.title(predicted_label_batch[n].title(), color=color)
  plt.axis('off')
_ = plt.suptitle("Model predictions (green: correct, red: incorrect)")
"""


label_id = val_data.classes
print("Predictions:", predicted_id)
print("Ground truth:", label_id)

cm = confusion_matrix(label_id, predicted_id)

plot_confusion_matrix(cm, class_names)
```
